```python
import matplotlib.pyplot as plt

# Server data
servers = [
    {"load": 3, "time": 0.315789},
    {"load": 5, "time": 0.526316},
    {"load": 8, "time": 0.842105},
    {"load": 2, "time": 0.210526},
    {"load": 7, "time": 0.736842}
]

# Extract load and time values
loads = [server["load"] for server in servers]
times = [server["time"] for server in servers]

# Plotting
plt.plot(loads, times, marker='o', linestyle='-')
plt.xlabel('Load')
plt.ylabel('Time (Seconds)')
plt.title('Load vs Time for Servers')
plt.grid(True)
plt.show()
```
